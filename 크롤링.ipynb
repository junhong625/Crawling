```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException, UnexpectedAlertPresentException
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait as WDW
from selenium.webdriver.support import expected_conditions as EC
import time

name_list = []
address_list = []
person_list = []
phonenumber_list = []

    
for idx in range(2,27):#2,27
    driver = webdriver.Chrome('chromedriver.exe') # ()안에 chromedriver.exe파일 경로 입력
    driver.get('http://www.nsdi.go.kr/lxportal/?menuno=4085')
    driver.find_element_by_xpath('//*[@id="shSido"]/option[2]').click() # 서울특별시
    time.sleep(1)
    driver.find_element_by_xpath('//*[@id="shSigungu"]/option[%s]'%(idx)).click() # 각 구
    time.sleep(1)
    element = WDW(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="icon_btn_write"]')))
    driver.find_element_by_xpath('//*[@id="icon_btn_write"]').click() #검색버튼 클릭
    time.sleep(1)
    full_html = driver.page_source
    soup = BeautifulSoup(full_html, 'html.parser')
    total = soup.select_one('#searchVO > div.bl_table > div.global-btn-wrap > p > span').get_text()
    if len(total) > 3:
        total = int(total.split(',')[0] + total.split(',')[1])
    if int(total) % 10 == 0:
        total = int(total)//10
    else:
        total = int(total)//10 + 1
    for t in range(total):
        full_html = driver.page_source
        soup = BeautifulSoup(full_html, 'html.parser')
        length = len(soup.select('.t_l'))//3 # len(soup.select('.t_l > a'))로 링크들을 계산하니 휴업으로 인해 링크가 없어 빠지게 되는 부분이 발생해 a를 빼고 3으로 나누어서 해결
        for n in range(length):
            try : # 처음엔 그냥 실행했으나 휴업으로 인해 링크가 설정되어 있지 않은 부분은 NoSuchElementException이 발생하기에 try-exception문으로 해결 
                element = WDW(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="searchVO"]/div[2]/div[2]/ol/li[12]/a')))
                driver.find_element_by_xpath('//*[@id="searchVO"]/div[2]/table/tbody/tr[%s]/td[3]/a'%(n+1)).click() # 상호명 링크 클릭
            except NoSuchElementException:
                full_html = driver.page_source
                soup = BeautifulSoup(full_html, 'html.parser')
                name_list.append(soup.select('.t_l')[n*3+1].get_text())
                address_list.append(soup.select('.t_l')[n*3+2].get_text())
                person_list.append(soup.select('.t_c')[n].get_text())
                phonenumber_list.append('휴업')
                continue
            try:
                full_html = driver.page_source
            except UnexpectedAlertPresentException:
                result = driver.switch_to_alert()
                result.accept()
                result.dismiss()
                full_html = driver.page_source
            soup = BeautifulSoup(full_html, 'html.parser')
            name = soup.select('.t_l')
            name_list.append(name[5].get_text())
            address_list.append(name[12].get_text())
            person_list.append(name[7].get_text())
            phonenumber_list.append(name[13].get_text())
            driver.back() 
        element = WDW(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="searchVO"]/div[2]/div[2]/ol/li[12]/a')))
        if t != total-1:
            driver.find_element_by_xpath('//*[@id="searchVO"]/div[2]/div[2]/ol/li[12]/a').click() #다음 번호 버튼 클릭
        else:
            driver.close()
```

```python
for t in range(total):
    full_html = driver.page_source
    soup = BeautifulSoup(full_html, 'html.parser')
    length = len(soup.select('.t_l'))//3 # len(soup.select('.t_l > a'))로 링크들을 계산하니 휴업으로 인해 링크가 없어 빠지게 되는 부분이 발생해 a를 빼고 3으로 나누어서 해결
    for n in range(length):
        try : # 처음엔 그냥 실행했으나 휴업으로 인해 링크가 설정되어 있지 않은 부분은 NoSuchElementException이 발생하기에 try-exception문으로 해결 
            element = WDW(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="searchVO"]/div[2]/div[2]/ol/li[12]/a')))
            driver.find_element_by_xpath('//*[@id="searchVO"]/div[2]/table/tbody/tr[%s]/td[3]/a'%(n+1)).click() # 상호명 링크 클릭
        except NoSuchElementException:
            full_html = driver.page_source
            soup = BeautifulSoup(full_html, 'html.parser')
            name_list.append(soup.select('.t_l')[n*3+1].get_text())
            address_list.append(soup.select('.t_l')[n*3+2].get_text())
            person_list.append(soup.select('.t_c')[n].get_text())
            phonenumber_list.append('휴업')
            continue
        try:
            full_html = driver.page_source
        except UnexpectedAlertPresentException:
            result = driver.switch_to_alert()
            result.accept()
            result.dismiss()
            full_html = driver.page_source
        soup = BeautifulSoup(full_html, 'html.parser')
        name = soup.select('.t_l')
        name_list2.append(name[5].get_text())
        address_list2.append(name[12].get_text())
        person_list2.append(name[7].get_text())
        phonenumber_list2.append(name[13].get_text())
        driver.back() 
    element = WDW(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="searchVO"]/div[2]/div[2]/ol/li[12]/a')))
    if t != total-1:
        driver.find_element_by_xpath('//*[@id="searchVO"]/div[2]/div[2]/ol/li[12]/a').click() #다음 번호 버튼 클릭
    else:
        driver.close()
```

```python
import pandas as pd
nl = pd.DataFrame(name_list, columns=['상호명'])
al = pd.DataFrame(address_list, columns=['소재지'])
pl = pd.DataFrame(person_list, columns=['대표자'])
phl = pd.DataFrame(phonenumber_list, columns=['전화번호'])
```

```python
Seoulsi = pd.concat([nl, al, pl, phl],axis=1)
```
